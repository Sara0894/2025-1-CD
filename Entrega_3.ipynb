# **Gráficos en Matplotilb**

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-colorblind')
```

**Gráfico de líneas**

```python
months = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
sales_product_x = [100, 120, 140, 160, 180, 200]
sales_product_y = [90, 110, 130, 145, 165, 185]

plt.figure(figsize=(8, 5))
plt.plot(months, sales_product_x, label='Producto X', marker='o')
plt.plot(months, sales_product_y, label='Producto Y', marker='s')
plt.title('Ventas mensuales por producto')
plt.xlabel('Mes')
plt.ylabel('Unidades vendidas')
plt.legend()
plt.grid(True)
plt.show()
```

**Gráfico de barras**

```python
categories = ['Alimentos', 'Tecnología', 'Hogar', 'Ropa']
sales= [230, 180, 150, 200]

plt.figure(figsize=(6, 4))
plt.bar(categories, sales, color='skyblue')
plt.title('Ventas por categoría')
plt.ylabel('Unidades')
plt.show()
```

**Gráfico de pastel**

```python
languages = ['Python', 'Java', 'C++', 'JavaScript']
percentages = [40, 25, 15, 20]

plt.figure(figsize=(6, 6))
plt.pie(percentages, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title('Preferencia de lenguajes de programación')
plt.axis('equal')
plt.show()
```

**Histograma**

```python
ages = [15, 22, 21, 23, 22, 19, 25, 30, 27, 22, 19, 21, 28, 30, 24, 19]

plt.figure(figsize=(7, 4))
plt.hist(ages, bins=6, color='orchid', edgecolor='black')
plt.title('Distribución de edades en una clase')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

**Dispersión**

```python
study_time = [1, 2, 3, 4, 5, 6, 7, 8]
qualification = [50, 55, 60, 70, 75, 80, 85, 90]

plt.figure(figsize=(7, 5))
plt.scatter(study_time, qualification, color='tomato')
plt.title('Relación entre horas de estudio y calificación')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificación obtenida')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
data = [
    {'Día': 'Lun', 'Ventas': 120, 'Zona': 'Norte', 'Cliente': 'A'},
    {'Día': 'Mar', 'Ventas': 150, 'Zona': 'Sur', 'Cliente': 'B'},
    {'Día': 'Mié', 'Ventas': 160, 'Zona': 'Norte', 'Cliente': 'C'},
    {'Día': 'Jue', 'Ventas': 140, 'Zona': 'Centro', 'Cliente': 'A'},
    {'Día': 'Vie', 'Ventas': 180, 'Zona': 'Sur', 'Cliente': 'C'},
    {'Día': 'Sáb', 'Ventas': 200, 'Zona': 'Centro', 'Cliente': 'B'},
    {'Día': 'Dom', 'Ventas': 170, 'Zona': 'Norte', 'Cliente': 'B'}
]

import pandas as pd
df = pd.DataFrame(data)
```

**Barplot**

```python
warm_colors = ['#FFE0B2', '#FFCC80', '#FFB74D', '#FFA726', '#FF8A65', '#F4511E', '#D84315']

plt.figure(figsize=(8, 5))
sns.barplot(x='Día', y='Ventas', data=df, hue='Día', palette=warm_colors, legend=False)
plt.title('Ventas diarias')
plt.show()
```

**Boxplot por zona**

```python
pink_colors = {    'Centro': '#F7C6D9',
    'Norte':  '#E75480',
    'Sur':    '#C71585'
                    }


plt.figure(figsize=(8, 5))
sns.boxplot(x='Zona', y='Ventas', data=df, hue='Zona', palette=pink_colors, legend=False)
plt.title('Distribución de ventas por zona')
plt.show()
```
